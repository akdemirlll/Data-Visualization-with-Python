# Laplacian of Gaussian # 
Bu Notebook çalışmasında, bilgisayarla görmede genellikle kenar algılama için kullanılan bir filtre olan Laplacian of Gaussian'ı (LoG) kullanacağız. Filtrenin kendisinin kullanımıyla ilgili çok fazla ayrıntıya değinmeyeceğiz. Bununla birlikte, belirli bir x ve y noktası için, değeri aşağıdaki formülle türetilir:

<img style="float: middle;" src="a.png">

LoG filtresinin tersini oluşturacağız. Birazdan göreceğiniz gibi, ilginç bir şekle sahip zarif bir matristir. Alternatif olarak, *axes3d.get_test_data* ile bazı test verileri alabilirsiniz. Ancak özel verilerle nasıl başa çıkılacağını göstermek daha eğlenceli olacaktır.

Bu, Gauss matrisinin Laplacian'ını oluşturmak için hızlı ve basit Python kodudur.

```python
def LoG(x, y, sigma):
    temp = (x ** 2 + y ** 2) / (2 * sigma ** 2)
    return -1 / (np.pi * sigma ** 4) * (1 - temp) * np.exp(-temp)
```

# Bazı Test Verileri Oluşturmamız Gerekiyor #
Şimdi bazı test verileri oluşturalım. Aşağıdaki tüm *X2*, *Y2*, *Z2*, 49 × 49 boyutunda 2d dizilerdir. Bu 2d dizilerini de 1d olarak yeniden şekillendiriyoruz, ör. *X1*, *Y1* ve *Z1*. Bu 1d dizileri daha sonra bazı grafikler çizmek için de kullanılacaktır.

```python
N = 49
half_N = N // 2
X2, Y2 = np.meshgrid(range(N), range(N))
Z2 = -LoG(X2 - half_N, Y2 - half_N, sigma=8)
X1 = np.reshape(X2, -1)
Y1 = np.reshape(Y2, -1)
Z1 = np.reshape(Z2, -1)
```

# Tel kafes #
Basitçe ızgara üzerindeki yakın noktaları birleştirerek en temel olay olan bir tel kafes çizimi ile başlayacağız.

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(X2, Y2, Z2, color='orange')
```

# Yüzey #
Bu, aynı zamanda çok açık olan bir yüzey grafiğidir. {Daha detaylı bilgi için kaynakça kısmında [1] şeklinde yazan yazının yanındaki linke basmanız yeterli} [1]

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X2, Y2, Z2, cmap='jet')
```

Yüzey grafiğinin dezavantajlarından biri, yüzey opak olduğu için verilerin bir kısmını karartacak olmasıdır. Peki renk haritalı bir tel kafesi yok mu ? Tabii ki var

```python
from matplotlib import cm
# Normalize the colors based on Z value
norm = plt.Normalize(Z2.min(), Z2.max())
colors = cm.jet(norm(Z2))
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X2, Y2, Z2, facecolors=colors, shade=False)
surf.set_facecolor((0,0,0,0))
```

Açıklamak gerekirse, bir tel kafes kutudan çıkar çıkmaz renk haritasını desteklemez. Bu nedenle, bunun yerine burada bir yüzey çiziyoruz. Ve sonra yüz renklerini kaldırırız, böylece geriye sadece kenar renkleri kalır.

# Çaydanlık #
Bilgisayar grafiklerinde, 3D alandaki herhangi bir nesne bir üçgen kümesine ayrıştırılabilir. Trisurf üçgenlemeyi desteklediğine göre, neden onunla bir 3D ağ dosyasından 3D bir nesne oluşturmuyorsunuz?
En temel 3D nesne dosyası biçimlerinden biri .obj'dir [3]. Bu dosya biçimine çok fazla girmeyeceğiz. Ancak özetle, noktalar ve yüzeyler için 3D konumları belirtir. Bu, bir .obj dosyasının pasajıdır:

```python
# List of geometric vertices
v 0.123 0.234 0.345
v ...
# Polygonal face element
f 1 2 3
f ...
```

V harfi veya f harfi ile bir satırın başına gelen köşeleri ve yüzleri belirtir. İşte bu dosya formatını okumak için kendi kendine hazırlanmış çok basit bir komut dosyası. Objenin desteklediği başka özellikler de vardır, ancak bunları basitçe görmezden geleceğiz.

```python
def read_obj(filename):
    triangles = []
    vertices = []
    with open(filename) as file:
        for line in file:
            components = line.strip(' \n').split(' ')
            if components[0] == "f": # face data
                # e.g. "f 1/1/1/ 2/2/2 3/3/3 4/4/4 ..."
                indices = list(map(lambda c: int(c.split('/')[0]) - 1, components[1:]))
                for i in range(0, len(indices) - 2):
                    triangles.append(indices[i: i+3])
            elif components[0] == "v": # vertex data
                # e.g. "v  30.2180 89.5757 -76.8089"
                vertex = list(map(lambda c: float(c), components[1:]))
                vertices.append(vertex)
    return np.array(vertices), np.array(triangles)
```

Şimdi tripsurf ile bir obj dosyası yükleyip oluşturalım. Aşağıdaki teapot.obj Stanford Computer Graphics Lab'den ödünç alınmıştır. Ancak bulabileceğiniz diğer tüm .obj dosyalarıyla deneyebilirsiniz.[4]

```python
vertices, triangles = read_obj("teapot.obj")
x = vertices[:,0]
y = vertices[:,1]
z = vertices[:,2]
ax = plt.axes(projection='3d')
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([0, 3])
ax.plot_trisurf(x, z, triangles, y, shade=True, color='white')
```
